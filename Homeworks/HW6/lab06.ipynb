## Завдання 2 

Переберіть всі підгрупи знакозмінної групи $A_5$. Перевірте всі на нормальність. Порівняйте час роботи з методом `.is_simple` для групи $A_5$. 

```python
import time
start_time = time.time()
A5 = AlternatingGroup(5)

def is_normal(subgroup, group)->bool: 
    rc = group.cosets(subgroup, side='right')
    lc = group.cosets(subgroup, side='left')
    rc_sorted = sorted ([sorted(coset) for coset in rc])
    lc_sorted = sorted ([sorted(coset) for coset in lc])
    return rc_sorted == lc_sorted

def simple_check(group)->bool:
    E = group.subgroups()[0]
    for subgroup in group.subgroups():
        if is_normal(subgroup, group) and subgroup != E and subgroup != group:
            return False
    return True

print(simple_check(A5))
end_time = time.time()
print(f"my simple check worked {end_time - start_time} secs")
```
